```python
S04_1 = pd.read_pickle('../an_data/EMG/S04_1.pkl')

S04_1['duration'] = S04_1.apply(lambda row: row.stop - row.start, axis=1)
```

```python
from datetime import datetime

ts = S04_1.iloc[0].start

type(ts)

datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S.%f')
```

```python
START_TS = datetime.fromtimestamp(S04_1.iloc[0].start)
S04_1['start_timestamp_raw'] = S04_1.apply(lambda row: datetime.fromtimestamp(row.start), axis=1)
S04_1['stop_timestamp_raw'] = S04_1.apply(lambda row: datetime.fromtimestamp(row.stop), axis=1)

S04_1['start_timestamp'] = S04_1.apply(lambda row: str(row.start_timestamp_raw - START_TS).replace('0 days', '').lstrip(), axis=1)
S04_1['stop_timestamp'] = S04_1.apply(lambda row: str(row.stop_timestamp_raw - START_TS).replace('0 days', '').lstrip(), axis=1)


S04_1['start_frame'] = S04_1.apply(lambda row: int((row.start_timestamp_raw - START_TS).total_seconds()*30 + 1), axis=1)
S04_1['stop_frame'] = S04_1.apply(lambda row: int((row.stop_timestamp_raw - START_TS).total_seconds()*30 + 1), axis=1)
```

```python
S04_1.head()
```

```python
VERBS = {verb: i for i, verb in enumerate(an_annotations_train.labels.unique().tolist())}
VERBS
```

```python

result = []
actual_uid = 0
for i in range(len(an_annotations_train)):
    row = an_annotations_train.iloc[i]
    if row.file == "S04_1.pkl":
        idx = row[0]
        sample = S04_1.iloc[idx]
        result.append({
            'uid': actual_uid,
            'start_ts': sample.start_timestamp,
            'end_ts': sample.stop_timestamp,
            'verb': row.labels,
            'verb_class': VERBS[row.labels],
            'description': sample.description,
            'start_frame': sample.start_frame,
            'stop_frame': sample.stop_frame,
            'subject': row.file.split('_')[0]
        })
        actual_uid += 1
        
result = pd.DataFrame(result)
result_test = []

for i in range(len(an_annotations_test)):
    row = an_annotations_test.iloc[i]
    if row.file == "S04_1.pkl":
        idx = row[0]
        sample = S04_1.iloc[idx]
        result_test.append({
            'uid': actual_uid,
            'start_ts': sample.start_timestamp,
            'end_ts': sample.stop_timestamp,
            'verb': row.labels,
            'verb_class': VERBS[row.labels],
            'description': sample.description,
            'start_frame': sample.start_frame,
            'stop_frame': sample.stop_frame,
            'subject': row.file.split('_')[0]
        })
        actual_uid += 1
        
result_test = pd.DataFrame(result_test)
```

```python
result
```

```python
import pickle

with open("train_val/S04_train.pkl", 'wb') as out:
    pickle.dump(result, out)

with open("train_val/S04_test.pkl", 'wb') as out:
    pickle.dump(result_test, out)
```
